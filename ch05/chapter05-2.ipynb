## salt & pepper 노이즈와 중간값 필터링

```python
#img = cv2.imread('/content/gdrive/My Drive/디지털 영상처리 강의 관련/Images/lena256.jpg')
img = cv2.imread('/content/gdrive/My Drive/디지털 영상처리 강의 관련/Images/fce(salt_pepper noise).bmp')

#print(img.shape)
#cv2_imshow(img)
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#plt.imshow(RGB_img)

gimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gimg.shape)
cv2_imshow(gimg)
plt.imshow(gimg, cmap='gray')
```

```python
from random import randint
```

```python
# salt & pepper noise
#10% noise

noiseRatio = 20
whiteMax = noiseRatio/2
blackMax = noiseRatio

noiseimg = gimg.copy()

for h in range(0, noiseimg.shape[0]):
    for w in range(0, noiseimg.shape[1]):
        val = randint(0, 100) # 0에서 99까지의 난수
        if val < whiteMax: # 5보다 작으면 화이트 노이즈
            noiseimg[h, w] = 255
        elif val < blackMax: # 5이상이고 10보다 작으면 블랙 노이즈
            noiseimg[h, w] = 0

cv2_imshow(noiseimg)
```

```python
out_img = cv2.medianBlur(noiseimg, 3)
cv2_imshow(out_img)

out_img2 = cv2.medianBlur(noiseimg, 5)
cv2_imshow(out_img2)
```
